```python
import pandas as pd
import codecs
```

```python
def get_metadata(image_file_name, char_main, file_handler_found=None, file_handler_not_found=None, meta_data=None):
    temp = char_main
    vowel_diacritic = ['া', 'ি', 'ী', 'ু', 'ূ', 'ৃ', 'ে', 'ৈ', 'ো', 'ৌ']
    consonant_diacritic = ['ঁ', 'র্', '্য', '্র', '্র্য', 'র্্র']
    store = temp
    for j in vowel_diacritic:
        if j in temp:
            search = meta_data[meta_data['char'] == j]
            vowel = search['label'].to_string(index=False)
            vowel = vowel.replace(" ", '')
            temp = temp.replace(j, '')
            break
        else:
            vowel = 0
    for j in consonant_diacritic:
        if j in temp:
            search = meta_data[meta_data['char'] == j]
            consonant = search['label'].to_string(index=False)
            consonant = consonant.replace(" ", '')
            temp = temp.replace(j, '')
            break
        else:
            consonant = 0
    temp = temp.replace('\n', '')
    temp = temp.rstrip()

    if temp:
        search = meta_data[meta_data['char'] == temp]
        root_char = search['label']
        if len(root_char):
            root_char = root_char.to_string(index=False)
        else:
            root_char = "Missing"
    else:
        if float(vowel) > 0 or float(consonant) > 0:
            root_char = 0
        else:
            root_char = "Missing"
    if root_char == 'Missing':
        print('Image name {}, Root char {} is missing with {} and {}'.format(image_file_name, temp, vowel, char_main))
        file_handler_not_found.write("{},{},{},{},{}".format(image_file_name, root_char, vowel, consonant, store))

    else:
        file_handler_found.write("{},{},{},{},{}".format(image_file_name, root_char, vowel, consonant, char_main))
```

```python
df = pd.read_csv('Meta_data/final_meta_old.csv')
filepath_found = "char_stat/found_19_oct.csv"
filepath_notfound = "char_stat/not_found_19_oct.csv"

with codecs.open('char_stat/ICT_Part1.tsv', encoding='utf-8') as f:
    with open(filepath_found, 'w', newline='', encoding="utf-8") as file_handler1:
        with open(filepath_notfound, 'w', newline='', encoding="utf-8") as file_handler:
            c = 0
            for line in f:
                if c > 0:
                    line = line.split('\t')
                    path = line[-2].split('/')    #for id path=line[1]
                    image_name = path[-1]
                    char = line[-1]
                    get_metadata(image_name, str(char), file_handler1, file_handler, df)
                c += 1
```

```python
from collections import Counter
import numpy as np
def char_frequency(char):
    char=list(char)
    c = Counter(char)
    c = c.most_common(len(c))
    c = dict(c)
    return c

def in_fre():
    df = pd.read_csv('char_stat/found_19_oct.csv',names=['image_id','grapheme_root','vowel_diacritic','consonant_diacritic','grapheme'])
    
    l1 =len(pd.get_dummies(df['grapheme_root']).values[0])
    group1 = df.groupby(['grapheme_root'])['image_id'].unique()
    
    m1 = np.unique(df['grapheme_root'])
    n1 = range(0,208)
    dif1 = np.setdiff1d(n1,m1)

    v = df[df['vowel_diacritic']!=0]
    l2 = len(v.vowel_diacritic.unique())
    m2 = v.vowel_diacritic.unique()
    n2 = range(1,11)
    dif2 = np.setdiff1d(n2,m2)
    group2 = v.groupby(['vowel_diacritic'])['image_id'].unique()
    
    c = df[df['consonant_diacritic']!=0]
    m3 = c.consonant_diacritic.unique()
    n3 = range(1,8)
    dif3 = np.setdiff1d(n3,m3)
    l3 = len(c.consonant_diacritic.unique())
    group3 = c.groupby(['consonant_diacritic'])['image_id'].unique()
    
    
    print('grapheme_root:',l1)
    print('vowel_diacritic:',l2)
    print('consonant_diacritic:',l3)
    print("Missing grapheme_root:",dif1)
    print("Missing vowel_diacritic:",dif2)
    print("Missing consonant_diacritic:",dif3)
```
